```python
def get_names(base_dir, subset, split):
    with open(f"{base_dir}/Chart-to-text-main/{subset}_dataset/dataset/dataset_split/{split}_index_mapping.csv", 'r') as f:
        split_names = f.readlines()
    
    split_names = split_names[1:]
    split_names = [x[:-1] for x in split_names]
    return split_names
    
```

```python
train_names, val_names, test_names = get_names(base_dir, "train"), get_names(base_dir, "val"), get_names(base_dir, "test")
```

```python
print(len(train_names), " @@ ", len(val_names), " ## ", len(test_names))
```

```python
subset = "pew"
twocol_names, multicol_names = os.listdir(f"{base_dir}/Chart-to-text-main/{subset}_dataset/dataset/captions"), os.listdir(f"{base_dir}/Chart-to-text-main/{subset}_dataset/dataset/multiColumn/captions")
```

```python
twocol_names, multicol_names = ['two_col-' + name for name in twocol_names], ['multi_col-' + name for name in multicol_names]
```

```python
pew_data = []
splits = [train_names, val_names, test_names]
for split_names in splits:
    split_data = {}
    for name in split_names:
        key = int('multi_col-' in name)
        address_map = {0: lambda x: f'{x}', 1: lambda x: f'multiColumn/{x}'}
        shutil.copy2(os.path.join(f"{base_dir}/Chart-to-text-main/{subset}_dataset/dataset/{address_map[key]('imgs')}",name.split('-')[1].split('.')[0]+'.png')
                    , f"{base_dir}/pew_images/{name}")
        with open(os.path.join(f"{base_dir}/Chart-to-text-main/{subset}_dataset/dataset/{address_map[key]('captions')}",name.split('-')[1]), 'r') as f:
            name_caption = f.read()
            split_data[name] = name_caption
    pew_data.append(split_data)
```
